### Reverse a String

```python
def reverseString (a_string: str) -> str : 
    """
    Function that takes a string (array) input and returns the string in reverse order
    This is O(n) time
    
    Args:
        a_string (str): Must be a string
        
    Return:
    """
    
    reversed_list = []
    
    for i in range(len(a_string) - 1, -1, -1):
        reversed_list.append(a_string[i])
    
    return ''.join(reversed_list)
    
    
    
reverseString("Hi My Name is Brad")
```

```python
def reverseStringSlicing (a_string: str) -> str : 
    """
    Function that takes a string (array) input and returns the string in reverse order
    This is O(n) time
    
    Args:
        a_string (str): Must be a string
        
    Return:
    """
    
    return a_string[::-1]

reverseStringSlicing("Hello")
```
